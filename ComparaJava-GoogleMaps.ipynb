```python
import  pandas, os
import time
from random import randint
import geopy.distance
import geopy
from geopy.geocoders import Nominatim
```

```python
df1=pandas.read_csv('COORD_JAVA_GMAPS.csv', sep=';', encoding='latin-1',decimal=',')
df1.head()
#df1.LAT_JAVA.apply(lambda x: x.str.replace(',','.'))
#df1.head()
```

```python
df4 = pandas.DataFrame(columns=['ID_FAMILIA_CADUNICO','DESVIO'])
retornos = 0
retate1km = 0
for index, row in df1.iterrows():
    try: 
        if row["LAT_JAVA"] != "None" and row["LAT_JAVA"] != "ERROR" and row["LAT_MAPS"] != "None" and row["LAT_MAPS"] != "ERROR":
            retornos=retornos + 1
            p1 = (float(row["LAT_JAVA"]),float(row["LONG_JAVA"]))
            p2 = (float(row["LAT_MAPS"]),float(row["LONG_MAPS"]))

            distkm = geopy.distance.distance(p1, p2).km
            df4.loc[index] = [row['ID_FAMILIA_CADUNICO'], distkm]
            if distkm <= 1:
                retate1km = retate1km + 1
            #print("%.3f" %(geopy.distance.distance(p1, p2).km))
    except:
    #else:
        df4.loc[index] = [row['ID_FAMILIA_CADUNICO'],"None"]
        #print(None)
```

```python
#df4
print(retornos)
print(retate1km)
```

```python
f=open("coordresultFinal.txt","w+")
f.write(df4.to_csv())
f.close()
```

```python
total = df4.shape[0]-1
print(total)
print(retornos/total)
print(retate1km/total)
f=open("estatisticasFinal.txt","w+")
f.write("Endereços pesquisados (sem pré-processamento): " + str(total) + "\n")
f.write("Número de retornos do OpenStreetMap: " + str(retornos) + " (" + str((retornos/total)*100) +"%)\n")
f.write("Número de retornos com desvio de até 1km: " + str(retate1km) + " (" +str((retate1km/total)*100) +"%)\n")
f.close()
```
